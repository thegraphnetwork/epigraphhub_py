# Downloading data from WorldPop.org

```python
import pandas as pd
from epigraphhub.data.worldpop import WorldPop
from IPython.display import display, Markdown, Image
```

To get started we instantiate a WorlPop object to Consume the API

```python
WP = WorldPop()
```

The representation of this object is a markdown table describing the contents of the API

```python
display(Markdown(f"{WP}"))
```

## Downloading table lists

Once we know which dataset we want to explore, for example, *population(pop)*, we can list all the tables available within it by organizational level:

```python
for df in WP.get_dataset_tables("pop"):
    display(Markdown(f"### {df['name']}: *{df['alias']}*"))
    display(df['data'])
```

We can dig deeper to get closer to the actual data: we pick an area, say *population (pop)*, a level, say *Unconstrained individual countries 2000-2020 ( 1km resolution ): "wpic1km"* and the country: *Guyana*.

```python
data = WP.get_data_by_country(alias='pop', level='wpic1km', ISO3_code='GUY')
data
```

The available file is a geotiff so we need to download it. First we can download a jpg thumbnail of the image we are trying to fetch:

```python
Image(data['data'][0]['url_img'])
```

```python
import rioxarray as rx
import wget
```

This geoTiff for Guyana has a resolution of 1 km per pixel

```python
wget.download(data['data'][0]['files'][0], out="guy_pop.tif")
```

```python
tif = rx.open_rasterio('guy_pop.tif')
tif
```

```python
tif.plot(figsize=(15,15));
```
