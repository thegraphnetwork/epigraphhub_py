```python
def search_in_database(search):
    '''
    Returns a dataframe with the database matched by full name or by partial name.
    
    :params search: string. Full name or keyword to search in the database
                            names. If the string 'all' is used in the parameter search all the 
                            databases names are returned. 
                            
    :returns: DataFrame.
    
    '''
    
    results = []
    
    if search.lower() == 'all':
        for source in wb.source.list():
            results.append(source)
            
    else: 
            
        byword = search.lower().split()

        search = search.replace(' ', '').lower()

        for source in wb.source.list():

            if source['name'].replace(' ','').lower() == search:


                results.append(source)

            else: 

                for i in byword:

                    if i in source['name'].lower():

                        results.append(source)
                    
    return pd.DataFrame.from_records(results)
```

```python
def search_in_indicators(search):
    
    '''
    Returns a dataframe with the indicators matched by partial name.
    
    :params search: string. keyword to search in the indicators name. 
    
    :returns: DataFrame 
    
    '''
    
    results = []
    
    for indicator in wb.series.list(q = search):
        results.append(indicator)
            
    
    return pd.DataFrame.from_records(results)


    
search_in_indicators('income') 
```

```python
def get_worldbank_data(ind, country,last_years):
    
    '''
    This function get a list of indicators according to some country from the world data bank 
    and return this series in a dataframe. 
    
    :params ind: List of strings|string. List with the indicators whose data you want to get.
    :params country: List of strings|string. List with the ISO-CODE for the countries whose 
                     data you want to get. 
    :params last_years: int. Number of last years whose you want to get
                         the population data.

    :returns: DataFrame 
    
    '''
    
    
    return wb.data.DataFrame(series=ind, economy=country, mrv=last_years, timeColumns= True)
    
```
